 Entender cómo se relacionan las escalas de temperatura es fundamental en las ciencias experimentales, y automatizarlo con Python es una excelente forma de aplicar el pensamiento computacional.Aquí tienes la guía completa solicitada para tu clase de Ciencias Naturales II.1. Fórmulas de ConversiónPara convertir entre estas dos escalas, utilizamos relaciones lineales basadas en los puntos de congelación y ebullición del agua.De Celsius a Fahrenheit:$$°F = (°C \times \frac{9}{5}) + 32$$(O también se puede usar $1.8$ en lugar de $9/5$)De Fahrenheit a Celsius:$$°C = \frac{(°F - 32) \times 5}{9}$$(O también se puede dividir entre $1.8$ después de restar $32$)2. Ejemplos Resueltos Paso a PasoEjemplo A: Convertir 25°C a Fahrenheit (Temperatura ambiente)Identificar datos: Tenemos $°C = 25$.Sustituir en la fórmula: $°F = (25 \times 1.8) + 32$.Multiplicar: $25 \times 1.8 = 45$.Sumar: $45 + 32 = 77$.Resultado: 25°C equivalen a 77°F.Ejemplo B: Convertir 98.6°F a Celsius (Temperatura corporal)Identificar datos: Tenemos $°F = 98.6$.Sustituir en la fórmula: $°C = (98.6 - 32) \times \frac{5}{9}$.Restar primero: $98.6 - 32 = 66.6$.Multiplicar por 5/9: $66.6 \times 0.5555... \approx 37$.Resultado: 98.6°F equivalen a 37°C.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngresa los grados Celsius a convertir: "))
    f = (c * 9/5) + 32
    print(f"---> {c}°C equivalen a {f:.2f}°F")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    c = (f - 32) * 5/9
    print(f"---> {f}°F equivalen a {c:.2f}°C")

def menu():
    """Controla el flujo principal del programa"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡Cepeda Darrell NL = 10!")
            break
        else:
            print("Opción no válida. Intenta de nuevo.")

# Llamada inicial para arrancar el programa
if __name__ == "__main__":
    menu()
```
